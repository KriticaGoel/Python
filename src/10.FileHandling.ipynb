### Reading the content from file

1. Open the file -> open()  -> open('example.txt','r')
2. First parameter file name
3. Second parameter mode
4. Read a file -> read()  -> file_data.read()
5. close the file -> close() ->file_data.close()
6. Write a file -> write() ->file_data.write("Writing a content in file")
7. Second parameter mode w/a -> w means overwrite existing content and a means write from end of file
8. with -> it used to open the file and no need to call close function ->with open('example.txt','r') as file_data:
9. write multiple lines -> writelines() -> argument is list of lines
10. Read binary files -> rb and wb
11. Read total line, total word, total character from a file -> readlines()
12. After writing moving the cursor to starting to read file --> seek(0)

#### Mode of file
1. r -> reading content from the file
2. w -> Writing content to the file
3. r+ -> first read then write
4. w+ -> first write then read
5. rb -> read binary file
6. wb -> write in binary file

```python
file_data = open('example.txt','r')
content = file_data.read()
print(content)
file_data.close()
```

```python
file_data = open('example.txt','w')
content = file_data.write("Writing a content in file")
file_data.close()
```

```python
file_data = open('example.txt','a')
content = file_data.write("\n Appending a content in file")
file_data.close()
```

```python
with open('example.txt','r') as file_data:
    content = file_data.read()
    print(content)
    
```

```python
lines =[' First Line\n','Second Line\n','Third Line\n']

with open('example.txt','a') as file_data:
    file_data.writelines(lines)
```

```python
data = b'\x00123x00456x00789'

with open('binary_file.txt','wb') as binary_data:
    binary_data.write(data)
```

```python
with open('binary_file.txt','rb') as binary_data:
    print(binary_data.read())
    
```

```python
with open('total_data.txt','r') as file_data:
    lines = file_data.readlines() # read line and store in list
    print(f"Total number of lines: {len(lines)}")
    total_words = sum(len(line.split(" ")) for line in lines)
    print(f"Total number of words: {total_words}")
    print(f"Total number of characters: {sum(len(line) for line in lines)}")


```

```python
with open('example.txt','w+') as file_data:
   file_data.write("This is a new line added to the file.")
   file_data.write("\nAdding another line.")
   file_data.seek(0)  # Move the cursor to the beginning of the file
   content = file_data.read()
   print(content)
   
```
