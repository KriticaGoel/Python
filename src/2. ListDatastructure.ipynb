#### Methods in List

1. **Append**- Append at end of list
2. **Insert** - insert at specific position
3. **Remove** - remove the first occurence element by value; Return None; O(n); ValueError if x not found
4. **Pop** - pop the value at given index; Return the removed element;O(1); IndexError in case of empty out of range
5. **Merging to List** - use keyword **extends** and add 
6. **index** - use to find the index of an element in list.
7. **count** - use to count number of time element present in list
8. **sort** - use to sort the list in ascending order
9. **reverse** - use to reverse the list
10. **clear** - use to remove all elements of list and make list empty
11. **enumerator** -getting index value along with actual value in a combination that we are sorting along the list
12. **len** - len(LIST) return number of element inside the list
