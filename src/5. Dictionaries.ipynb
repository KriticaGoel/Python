```python
students ={"student_1":{"name":"Aung","age":25,"city":"Mandalay"},
           "student_2":{"name":"Kyi","age":22,"city":"Yangon"},
           "student_3":{"name":"Laavanya","age":23,"city":"Naypyitaw"}}
```

```python
for student_key,student_value in students.items():
    print( f"key is {student_key} and value is {student_value}")
    for data_key,data_value in student_value.items():
        print(f"    {data_key} : {data_value}")
```
