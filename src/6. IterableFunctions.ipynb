```python
# Strong password validator
# length>8 character
# contains at least one uppercase letter
# contains at least one lowercase letter
# contains at least one digit
# contains at least one special character from [@,#,$,%,&,*]


def is_strong_password(password):
    if len(password)<=8:
        return False
    has_upper=False
    has_lower=False
    has_digit=False
    has_special=False
    special_characters = "@#$%&*"
    for char in password:
        if(char.isupper()):
            has_upper=True
        elif(char.islower()):
            has_lower=True
        elif(char.isdigit()):
            has_digit=True
        elif(char in special_characters):
            has_special=True
    return has_upper and has_lower and has_digit and has_special
```

```python
print(is_strong_password("pass"))
```

```python
print(is_strong_password("Passw0rd@"))
```

```python
def passowrd_strength(password):
    if(len(password)<=8):
        return "Weak"
    if not any (char.isupper() for char in password):
        return "Uppercase Missing"
    if not any(char.islower() for char in password):
        return "Lowercase Missing"
    if not any(char.isdigit() for char in password):
        return "Digit Missing"
    if not any(char in "@#$%&*" for char in password):
        return "Special Character Missing"
    return "Strong Password"
```

```python
print(passowrd_strength("Passw0rd@"))
```

```python
#Palindorme checker
def is_palindrome(s):
    s= s.lower().replace(" ","")
    
    if s == s[::-1]:
        return True
    return False

print(is_palindrome("a sdadba"))
```

```python
# Optimized Palindrome Checker - Two Pointer Approach
def is_palindrome_optimized(s):
    left = 0
    right = len(s) - 1
    
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    
    return True

# Test cases
print(is_palindrome_optimized("mabam"))      # True
print(is_palindrome_optimized("hello"))      # False
print(is_palindrome_optimized("racecar"))    # True
print(is_palindrome_optimized("a"))          # True
print(is_palindrome_optimized("ab"))         # False
```
