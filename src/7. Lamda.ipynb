#### One line function defined by Lamda kyword

Syntax

 lamda arguments:experssion

 lamda -> keyword that tell interpretator that its a lamda function

 arguments -> Input for our function

 expression -> logic (function)

```python
sq=lambda x : x**2
print(sq(5))
```

```python
sum = lambda x,y:x+y
sum(100,200)
```
