#### Map fucntion help us in avoiding unnecessary line on code and we can apply the operation on all values or items in an iterable

### Syntax of Map()

list(map(function,iterable))


```python
print(list(map(lambda x:x**2,[1,2,3,4,5])))
```

```python
abc=[1,2,3,4,5]
print(list(map(lambda x:x**2,abc)))
```

```python
def sq(x):
    return x**2

abc=[1,2,3,4,5]
print(list(map(sq,abc)))
```

```python
print(list(map(lambda x,y:x+y, [1,2,3],[4,5,6])))
```

```python
print(list(map(int,['1','2','3','4','5'])))
```

```python
print(list(map(str.upper,['apple','banana','cherry'])))
```
