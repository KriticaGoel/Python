### task 3 Product Pricing

```python
#1.
price_dict ={"Shirt":400.00,"Pants":520.00,"Socks":80.00,"Jacket":1000.00,"Belt":50.50,"Hat":100.00}
```

```python
#2.
price_dict["Shoes"]=750.00
price_dict["Pants"]=700.00
removed_price = price_dict.pop("Shoes", "Item not found")
price_dict
```

```python
#3.
avgPrice=0.00
for item, price in price_dict.items():
    avgPrice += price
avgPrice /= len(price_dict)
print(f"Average Price: {avgPrice}")
```

```python
# Extra
max_price_item=""
min_price_item=""
for item,price in price_dict.items():
    if price == max(price_dict.values()):
        max_price_item=item
    if price == min(price_dict.values()):
        min_price_item=item

print(f"Highest Priced Item: {max_price_item} at {price_dict[max_price_item]}")
print(f"Lowest Priced Item: {min_price_item} at {price_dict[min_price_item]}")
```

### Task 4: Combined Operation

```python
#1.
catalog = []
for product in products:
    price = price_dict.get(product, 0.00)
    category = "Apparel" if product in ["Shirt", "Pants", "Socks", "Hat", "Jacket", "Belts", "Gloves"] else "Footwear"
    catalog.append((product, price, category))
catalog
```

```python
#2.
category_to_products ={}
for item,price,category in catalog:
    if(category not in category_to_products.keys()):
        category_to_products[category]=[item]
    else:
        category_to_products[category] +=[ item]
```

```python
#3.
max_products=0
max_category=""

for category,items in category_to_products.items():
    if(len(items)>max_products):
        max_products=len(items)
        max_category=category
print(category_to_products.get(max_category))
```
