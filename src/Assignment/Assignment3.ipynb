#### Task 1: Basic function : Price after discount

```python
def apply_discount(price,discount_percentage=5):
    if(discount_percentage<0 or discount_percentage>60):
        print("Invalid discount percentage")
        return price
    return price - (price * discount_percentage / 100)


apply_discount(1000,10)
apply_discount(500)
```

#### Task 2: Recursive function : Factorial Utility

```python
def factorial(n):
    if(n<0):
        return "Invalid input"
    if n==0 or n==1:
        return 1
    else:
        return n*factorial(n-1)
factorial(5)
factorial(0)
factorial(-3)
```

#### Task 6: Combined Utility Function

```python
def process_prices(prices):
    # Apply a 10% discount to each price
    discounted_prices = list(map(lambda price: price - (price * 0.10), prices))
    # Filter discounted price>300
    filtered_prices = list(filter(lambda price: price > 300, discounted_prices))
    return discounted_prices,filtered_prices

prices = [100, 500,900,50,750]
process_prices(prices)
```

#### Task 7 : Mini Problem: Menu using function

```python
def add_price(prices_list,price):
    return list(map(lambda p: p + price, prices_list))

def get_average_price(prices_list):
    return sum(prices_list)/len(prices_list) if prices_list else 0

def get_max_price(prices_list):
    return max(prices_list) if prices_list else None


user_input = input("Enter prices separated by commas:")
if(user_input.strip() == ""):
    print("No prices entered. Exiting.")
    exit()

elif not all(item.isdigit() for item in user_input.split(',')):
    print("Invalid number format. Exiting.")
    exit()
else:
    prices_list = list(map(float, user_input.split(',')))
    print("Choose any one option from below:")
    print("1 -> add price to each item")
    print("2 -> show average price")
    print("3 -> Show Highest price")
    print("q -> quit")

    while True:
        choice = input("Enter your choice 1/2/3/q:")
        if choice == '1':
            add_value = float(input("Enter price to add:"))
            updated_prices = add_price(prices_list, add_value)
            print("Updated Prices:", updated_prices)
        elif choice == '2':
            average_price = get_average_price(prices_list)
            print("Average Price:", average_price)
        elif choice == '3':
            max_price = get_max_price(prices_list)
            print("Highest Price:", max_price)
        elif choice.lower() == 'q':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please try again.")
```
